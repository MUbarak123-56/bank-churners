```python
import seaborn as sns
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/BankChurners.csv")
```

```python
df.columns
```

```python
df = df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
       'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis = 1)
df.head()
```

```python
df.to_csv("data/bankchurnersuse.csv", index = False)
```

### Distribution of Total Revolving Balance

```python
plt.figure(figsize= (20,6))
sns.distplot(a = df["Total_Revolving_Bal"], hist = False)
plt.title("Distribution of Total Revolving Balance")
plt.xlabel("Total Revolving Balance")
plt.ylabel("Frequency")
plt.show()
```

### Distribution of Credit Limit w/ Histogram

```python
plt.figure(figsize= (20,6))
sns.distplot(a = df["Credit_Limit"], hist = True)
plt.title("Distribution of Credit Limit")
plt.xlabel("Credit Limit")
plt.ylabel("Frequency")
plt.show()
```

### Credit Limit w.r.t Churn

```python
plt.figure(figsize= (10,6))
sns.boxplot(x = "Attrition_Flag", y = "Credit_Limit", data = df)
plt.title("Distribution of Credit Limit w.r.t Churn")
plt.xlabel("Attrition Flag")
plt.ylabel("Credit Limit")
plt.show()
```

### Total Revolving Balance w.r.t Churn

```python
plt.figure(figsize= (10,6))
sns.violinplot(y = "Total_Revolving_Bal", x = "Attrition_Flag", data = df, hue = "Attrition_Flag")
#plt.scatter(x = df["Credit_Limit"], y = df["Total_Revolving_Bal"],  = df["Attrition_Flag"])
plt.title("Total Revolving Balance w.r.t Churn")
plt.xlabel("Attrition")
plt.ylabel("Total Revolving Balance")
plt.show()
```

### Credit Limit w.r.t Education Level

```python
plt.figure(figsize= (20,10))
sns.boxplot(x = "Education_Level", y = "Credit_Limit", data = df, hue = "Attrition_Flag")
plt.title("Distribution of Credit Limit w.r.t Education")
plt.xlabel("Attrition Flag")
plt.ylabel("Education Level")
plt.show()
```

### Credit Limit w.r.t Marital Status

```python
plt.figure(figsize= (20,10))
sns.boxplot(x = "Marital_Status", y = "Credit_Limit", data = df, hue = "Attrition_Flag")
plt.title("Distribution of Credit Limit w.r.t Marital Status")
plt.xlabel("Marital Status")
plt.ylabel("Credit Limit")
plt.show()
```

### Total Revolving Balance w.r.t Marital Status

```python
plt.figure(figsize= (20,10))
sns.violinplot(x = "Marital_Status", y = "Total_Revolving_Bal", data = df, hue = "Attrition_Flag")
plt.title("Distribution of Total Revolving Balance w.r.t Marital Status", fontsize = 18)
plt.xlabel("Marital Status", fontsize = 12)
plt.ylabel("Total Revolving Balance", fontsize = 12)
plt.show()
```
